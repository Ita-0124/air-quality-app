```python
url = "https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=9e565f9a-84dd-4e79-9097-d403cae1ea75&limit=1000&sort=ImportDate desc&format=JSON"
hisurl = "https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=9b651a1b-0732-418e-b4e9-e784417cadef&limit=1000&sort=datacreationdate desc&format=JSON"
with requests.get(url) as response:
    response.encoding = "utf-8"
    data = response.json()
#print(data)
with requests.get(hisurl) as response:
    response.encoding = "utf-8"
    hisdata = response.json()
#print(hisdata)
```

```python
records_data = data['records']
records_hisdata = hisdata['records']
print('共有{}筆資料,紀錄時間:{}'.format(len(records_data),records_data[0]['publishtime']))
print('共有{}筆資料,最新一筆紀錄時間:{}'.format(len(records_hisdata),records_hisdata[0]['datacreationdate']))
```

```python
for d in records_data:    
    print('{},{}: AQI= {},status={}'.format(d['county'],d['sitename'],d['aqi'],d['status']))
```

```python
for d in records_hisdata:    
    print('紀錄時間:{} {},{}: AQI= {},status={}'.format(d['datacreationdate'],d['county'],d['sitename'],d['aqi'],d['status']))
```

```python
all_times = [datetime.strptime(d['datacreationdate'], '%Y-%m-%d %H:%M') for d in records_hisdata]

min_time = min(all_times)
max_time = max(all_times)

#unique_times = sorted(set(all_times), reverse=True)
#print("包含的時間點：")
#for t in unique_times:
#    print(t.strftime('%Y-%m-%d %H:%M'))


print(f"資料時間範圍：{min_time.strftime('%Y-%m-%d %H:%M')} 到 {max_time.strftime('%Y-%m-%d %H:%M')}\n")

time_counts = Counter(all_times)

print("每個時間點的資料數量：")
for time_point in sorted(time_counts, reverse=True):
    print(f"{time_point.strftime('%Y-%m-%d %H:%M')} ： {time_counts[time_point]} 筆")
```
